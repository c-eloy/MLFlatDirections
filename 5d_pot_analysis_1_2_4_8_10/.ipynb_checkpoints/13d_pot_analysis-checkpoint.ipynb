```python
# compute (||∇V||^2)
def grad_norm_squared(V, x ,factor=1):
    with tf.GradientTape() as tape:
        tape.watch(x)
        loss = V(x)  # compute V at each point
    gradients = tape.gradient(loss, x)  # compute ∇V at each point
    norm_squared = tf.reduce_sum(gradients**2, axis=1)  # ||∇V||^2 at each point
    return norm_squared * factor

def get_grad(V, x):
    with tf.GradientTape() as tape:
        tape.watch(x)
        loss = V(x)  # compute V at each point
    gradients = tape.gradient(loss, x)  # compute ∇V at each point
    return gradients
```

```python
####### Gradient Descent #######
```

```python
# randomly initialize the points
n_points = 100000
x_3d = tf.Variable(2*np.random.rand(n_points, 3)-1, dtype=tf.float32)

history_3d = []
```

```python
# Oprimizer for the gradient descent

init_alpha=10**(-4)
alpha = init_alpha

optimizer = tf.optimizers.Adam(learning_rate=alpha)
#optimizer = tf.optimizers.SGD(learning_rate=init_alpha,momentum=0.0)

# Learning loop 
loss_prev_step=10**8

#update_alpha = False
update_opt = False 
for step in range(500):  
    with tf.GradientTape() as tape:
        loss = tf.reduce_sum(grad_norm_squared(pot.V3d_sugra, x_3d)) # minimize ||∇V||^2 for all points
    
    # Compute ||∇V||^2 with respect to x
    gradients = tape.gradient(loss, [x_3d])

    #if update_alpha and step % 300 == 0 and step != 0:
    #    alpha = alpha/2
    #    optimizer = tf.optimizers.Adam(learning_rate=alpha)
    if update_opt and step % 200 ==0 :
        print("Reinitializing the optimizer")
        optimizer = tf.optimizers.Adam(learning_rate=init_alpha)
    
    if step == 700:
        update_opt = False 
        optimizer.learning_rate.assign(0.0001)
    
    # Apply gradient descent for all points
    optimizer.apply_gradients(zip(gradients, [x_3d]))
    
    history_3d.append(loss)
    
    if step % 10 == 0:
        print(f"""Step {step}: ||∇V||^2 = {loss.numpy()} with learning rate {optimizer.get_config()["learning_rate"]}""")
        
    if np.log(loss.numpy())/np.log(10)<-10: 
        print("Converged enough") 
        print(f"Step {step}: ||∇V||^2 = {loss.numpy()}")
        break
```

```python
##### Auto encoder analysis #####
```

```python
set_train, set_test = train_test_split(x_3d.numpy(), test_size=0.1, random_state=1)
```

```python
history_all = []

for i in range(3):
    
    this_latent_dim = i + 1
    input_dim = 3
    
    print(this_latent_dim)

    inputs = layers.Input(shape=(input_dim,))
    encoded = layers.Dense(64, activation='sigmoid')(inputs)
    encoded = layers.Dense(64, activation='sigmoid')(encoded)
    encoded = layers.Dense(this_latent_dim)(encoded)  

    decoded = layers.Dense(64, activation='sigmoid')(encoded)
    decoded = layers.Dense(64, activation='sigmoid')(decoded)
    decoded = layers.Dense(input_dim)(decoded)  

    this_autoencoder = keras.Model(inputs, decoded)
    this_autoencoder.compile(optimizer='adam', loss='mse')
    
    history_all.append(this_autoencoder.fit(set_train, set_train, epochs=200, batch_size=1000, shuffle=True, validation_data=(set_test,set_test)))
```

```python
history_mean_val = []
history_mean = []
for this_history in history_all: 
    this_mean_val = np.sum(this_history.history['val_loss'][-20:])/20
    this_mean = np.sum(this_history.history['loss'][-20:])/20
    history_mean_val.append(this_mean_val)
    history_mean.append(this_mean)
```

```python
plt.figure(figsize=(7,7))
plt.scatter(np.linspace(1,3,3),np.log(history_mean))
plt.scatter(np.linspace(1,3,3),np.log(history_mean_val))
plt.show()
```
